## Understanding Oscillations & Waves

In this session we want to gain an intuition around oscillations and waves. In mathematics and physics, it is often useful to have an ansatz when attempting to fit curves to reduce the numerical cost when finding a solution. In signal processing, for example, it is useful to have a working understanding of "wavepackets" which, rather than the usual single characteristic frequency, will have two or more.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Sinusoidal Waves

We start with basic waves of the form $y=\sin kx$. Note what happens as $k$ varies -- can you find a relationship between the wavenumber $k$ and the wavelength $\lambda$?

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y = np.sin(x)

plt.plot(x,y)
```

### Decaying Waves

We now look at waves of the form $y=e^{-\Lambda x}\sin kx$. Note what happens when you vary the decay constant $\Lambda$ and the wavenumber $k$.

As an extention, verify algebraically (i.e., with pen and paper/not numerically) $\frac{d^2y}{dx^2}+(k^2-\Lambda^2)y=-2\Lambda ke^{-\Lambda x}\cos kx$.

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y = np.exp(-x)*np.sin(x)

plt.plot(x,y)
```

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y = np.exp(-x)*np.sin(np.pi*x)

plt.plot(x,y)
```

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y = np.exp(-2*x)*np.sin(x)

plt.plot(x,y)
```

### Wavepackets

We now consider wavepackets and showing how they can be enveloped, acting as "waves within waves".

First let us consider what happens when we add sinusoids of two kinds.

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y = np.sin(x)+np.cos(x)

plt.plot(x,y)
```

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y = np.sin(x)+3*np.cos(x)

plt.plot(x,y)
```

This could be useful to know, however we haven't retrieved wavepackets. We shall look at products of sinusoids instead, and see if this improves the situation.

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y = np.sin(16*x)*np.sin(x)

plt.plot(x,y)
```

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y1 = np.sin(16*x)*np.sin(x)
y2 = np.sin(x)

plt.plot(x,y1)
plt.plot(x,y2,linestyle="--")
```

```python
a = float(input("What is the minimum x value in your domain? "))
b = float(input("What is the maximum x value in your domain? "))
x = np.linspace(a,b,1000)
y1 = np.sin(16*x)*np.sin(x)
y2 = np.sin(x)

plt.plot(x,y1)
plt.plot(x,y2,"--")
```
